```python
import pyautogui as pag
import mss, cv2
import numpy as np

pag.PAUSE = 0.08  # 클릭 후 대기 시간

seven_icon_postion = {'left': 43, 'top': 336, 'width': 15, 'height': 15}
three_icon_postion = {'left': 236, 'top': 444, 'width': 15, 'height': 15}

plus_button = [337, 452]
calc_button = [337, 500]

########## 이미지 판별 함수
def what_kind_of_image(img):
    mean = np.mean(img, axis=(0, 1)) # 3*n 다차원 행렬에서 각 행렬의 Xij 값을 더하여 평균을 낸다. 즉 결과는 3*1로 나온다. 

    result = "UNKNOWN"
    if (201 < mean[0]) and (mean[0] < 212) and (201 < mean[1]) and (mean[0] < 212) and (201 < mean[2]) and (mean[2] < 212):
        result = '7'
    elif (190 < mean[0]) and (mean[0] < 192) and (190 < mean[1]) and (mean[0] < 192) and (190 < mean[2]) and (mean[2] < 192):
        result = '3'

    return result

########## 정의된 좌표의 이미지 획득 함수
def get_target_image():
    with mss.mss() as sct:
        seven_img = sct.grab(seven_icon_postion)
        seven_img = np.array(seven_img)
        seven_img = seven_img[:, :, :3]

        three_img = sct.grab(three_icon_postion)
        three_img = np.array(three_img)
        three_img = three_img[:, :, :3]
    return (seven_img, three_img)

######## 이미지 미리보기
# cv2.imshow('seven_img', seven_img)
# cv2.imshow('three_img', three_img)
# cv2.waitKey(0)

## 지정된 좌표를 클릭하는 함수
def mouse_click(coords):
   pag.moveTo(x=coords[0], y=coords[1], duration=0.0)
   pag.mouseDown()
   pag.mouseUp()


is_found_seven = False
is_found_three = False
while True:
######### 죄표 얻어오기
#    x, y = pag.position()
#    position_xy = 'X: ' + str(x) + '   ' + 'Y: ' + str(y)
#    print (position_xy)

    ########## 정의된 좌표에서 이미지를 얻어 온다.
    seven_img, three_img = get_target_image()

    ########## 이미지를 판별한다.
    is_seven_img = what_kind_of_image(seven_img)
    is_three_img = what_kind_of_image(three_img)

    ### 판별된 이미지가 7이면
    if is_seven_img == '7' and not is_found_seven:
        is_found_seven = True
        mouse_click([seven_icon_postion['left'], seven_icon_postion['top']])
        print('Found 7............')

    ### 7이 판별되었고, 판별된 이미지가 3이면
    if is_found_seven and is_three_img == '3' and not is_found_three:
        is_found_three = True
        mouse_click(plus_button)
        print('Plus Click............')
        mouse_click([three_icon_postion['left'], three_icon_postion['top']])
        print('Found 3............')

    ### 둘 다 찾아서 클릭했으면 더하기 클릭 후 계산 버튼을 클릭한다.
    if is_found_seven and is_found_three:
        mouse_click(calc_button)
        print("계산완료 !!!")
        break

```
